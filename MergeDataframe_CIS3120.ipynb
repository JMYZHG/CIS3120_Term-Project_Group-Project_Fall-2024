```python
import pandas as pd
from nba_api.stats.static import players  #Getting active players
from nba_api.stats.endpoints import commonplayerinfo
csv_file_path = '/content/final_nba_data.csv' #Loading the CSV file
df_csv = pd.read_csv(csv_file_path)
df_csv['2024/2025'] = df_csv['2024/2025'].replace({'\$': '', ',': ''}, regex=True).astype(float) #Cleaning the data (removeing the $, commas)

active_players = players.get_active_players() #Getting active NBA players using nba_api
df_active_players = pd.DataFrame(active_players)

df_active_players = df_active_players.rename(columns={'full_name': 'player'}) #renaming the colmuns
merged_df = pd.merge(df_csv, df_active_players, on='player', how='inner') #merging two datasets on player colmun

print("The Merged DataFrame:") #showing combined data
print(merged_df.head())

if '2024/2025' in merged_df.columns:  #comparing salaries
    print()
    print()
    print("\nSalary Information Of The NBA PLayers (2024/2025):")
    print(merged_df[['player', '2024/2025']])
else:
    print("\n** No Salary Information Found (In The Combined/Merged Data) **.")
```
